```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices1.csv")
df
```

```python
%matplotlib inline
```

```python
plt.xlabel('area(sqr ftr)')
plt.ylabel('price(us $)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
```

```python
# Now we are going to make the model train 
```

```python
new_df = df.drop('price', axis = 'columns')
new_df
```

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
# Now we are make an linear object which we are going to train 
reg.fit(new_df, price)
```

```python
reg.predict([[3000]])
```
