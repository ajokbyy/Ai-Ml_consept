```python
# ============================***=======================================
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent(x, y):
    m_curr = 0
    b_curr = 0
    rate = 0.01
    n = len(x)

    # scatter actual points
    plt.scatter(x, y, color='red', marker='+', linewidth=3)

    for i in range(10000):
        y_predicted = m_curr * x + b_curr

        # plot line every 1000 iterations
        if i % 1000 == 0:
            plt.plot(x, y_predicted, color='green', alpha=0.4)

        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)

        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * bd

    # final best-fit line (bold)
    plt.plot(x, m_curr * x + b_curr, color='green', linewidth=3)
    plt.show()

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x, y)
```
