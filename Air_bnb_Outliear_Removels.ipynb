```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("AB_NYC_2019.csv")
df1
```

```python
df1.describe()
```

```python
df1.price.quantile(0.9) #np.float64(269.0)
df1.price.quantile(0.1)# np.float64(49.0)
```

```python
#So we are going remove some outliear whuch are 0 percential and 100 persecntial
percential_99 = df1.price.quantile(0.9)
percential_99

percential_1 = df1.price.quantile(0.1)
percential_1
```

```python
# now we are make a data frame with value from 49 to 269

df_no_outlier = df1[(df1.price<=percential_99)& (percential_1<=df1.price) ]
df_no_outlier
```
